```python
import pandas as pd

url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)

# print(df.head())
print(df.info())
```

```python
# Convert 'TotalCharges' to numeric, coercing errors to NaN
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')

mean_value = df["TotalCharges"].mean()

mode_value = df["SeniorCitizen"].mode()

median_value = df["MonthlyCharges"].median()

print("Total Charges: ",mean_value)
print("Senior Citizen: ",mode_value)
print("Monthly Charges: ",median_value)
```

```python
null_counts = df['TotalCharges'].isnull().sum()
print("Null values in 'TotalCharges' column:")
print(null_counts)
```

```python
# cleaning the dataset:

# removing null values:
df.dropna(inplace=True)

# replacing null values:
# df.fillna(value=0, inplace=True)

# removing duplicates:
df.drop_duplicates(inplace=True)
```

```python
df.Churn.value_counts()
```

```python
# axis represents rows or columns, 0 represents rows, 1 represents columns
df.drop(["Contract", "PaymentMethod", "InternetService"], axis = 1, inplace=True)
print(df.head())
```

```python
df.dropna(subset=['TotalCharges'], inplace=True)
```

```python
#converting texts to numbers so that machine can understand
df = pd.get_dummies(df, drop_first=True)
print(df.head)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop("Churn_Yes", axis = 1)
y = df["Churn_Yes"]

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# test size indicates the split of test and train, how much % should go to test.
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Precision:", precision_score(Y_test, y_pred))
print("Recall:", recall_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("Classification Report:\n", classification_report(Y_test, y_pred))

```
